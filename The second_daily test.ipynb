### 1、please import pandas 

```python
import pandas as pd
```

### 2、please read the Breast_cancer_data_for_ML.csv file and saved as DataFrame and name it as 'data'.

```python
df = pd.read_csv("Breast_cancer_data_for_ML.csv", index_col = 'id') #32
df
```

### 3、please drop the 'Unnamed: 32' column

```python
df.drop(['Unnamed: 32'],axis = 'columns' ,inplace = True)
```

### 4、please to use 1,0 instead of 'M' and 'B'

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B':0})
```

### 5、please split the data into training data and test data

```python
from sklearn.model_selection import train_test_split
y = df['diagnosis']
X=df.drop('diagnosis',axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

### 6、please do data preprocesing 

```python
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
pca = decomposition.PCA(n_components=2)

X_pca_scaled = pca.fit_transform(X_scaled)

print('Projecting %d-dimensional data to 2D' % X_scaled.shape[1])
```

### 7、please use SVM to do classification

```python
from sklearn.svm import SVC
name_model=SVC()
SVM_model = SVC(
    C=1.0,
    kernel='rbf',
    degree=3,
    gamma='auto',#'auto_deprecated',
    coef0=0.0,
    shrinking=True,
    probability=False,
    tol=0.001,
    cache_size=200,
    class_weight=None,
    verbose=False,
    max_iter=-1,
    decision_function_shape='ovr',
    random_state=None,
)
SVM_model.fit(X_train,y_train)
```
